```python
from transformers import pipeline, AutoTokenizer, AutoModelForTokenClassification, AutoConfig
# from transformers import *
import torch
import tensorflow
```

```python
tokenizer = AutoTokenizer.from_pretrained("mrm8488/codebert-base-finetuned-stackoverflow-ner")
model = AutoModelForTokenClassification.from_pretrained("mrm8488/codebert-base-finetuned-stackoverflow-ner")
nerpipeline = pipeline(model="mrm8488/codebert-base-finetuned-stackoverflow-ner")
```

```python
tokens = [tokenizer.cls_token] + tokenizer.tokenize("VScodeServer") + [tokenizer.sep_token]
token_ids = tokenizer.convert_tokens_to_ids(tokens)
nerpipeline = pipeline('ner', model=model, tokenizer=tokenizer)
```

```python
def get_context_embeddding(tokenizer, model, input_str):
    # tokens = [tokenizer.cls_token] + tokenizer.tokenize(input_str) + [tokenizer.sep_token]
    tokens = tokenizer.tokenize(input_str)
    print(tokens)
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    context_embeddings=model(torch.tensor(token_ids)[None,:])[0]
    return context_embeddings
    
```

```python
sentences = [
    "SQL table name",
    "select column in row",
    "delete node linked list",
    "a cat playing with yarn",
    "yarn cluster management"
]
```

```python
context_embeddings = [get_context_embeddding(tokenizer, model, sentence) for sentence in sentences]
```

```python
first_vectors = [embedding.detach().numpy()[0, 0, :] for embedding in context_embeddings]
```
