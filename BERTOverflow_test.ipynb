```python
def get_context_embeddding(tokenizer, model, input_str):
    # tokens = [tokenizer.cls_token] + tokenizer.tokenize(input_str) + [tokenizer.sep_token]
    tokens = tokenizer.tokenize(input_str)
    print(tokens)
    token_ids = tokenizer.convert_tokens_to_ids(tokens)
    context_embeddings=model(torch.tensor(token_ids)[None,:])[0]
    return context_embeddings
    
```

```python
sentences = [
    "SQL table name",
    "select column in row",
    "delete node linked list",
    "a cat playing with yarn",
    "yarn cluster management"
]
```

```python
context_embeddings = [get_context_embeddding(tokenizer, model, sentence) for sentence in sentences]
```
