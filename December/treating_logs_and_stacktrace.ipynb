```python
from gensim.models import Word2Vec
import gensim
from nltk.tokenize import sent_tokenize, word_tokenize
import warnings
import sys; sys.path.insert(0, '..')
import os
import my_utils
```

```python

for table_name in my_utils.table_names:
    # data stores the token of each bug report desc
    data = []
    folder_path = "./{}".format(table_name)
    if not os.path.exists(folder_path):
        print("{} does not exist".format(folder_path))
        continue
    
    logStacktrace_path = os.path.join(folder_path, "logStackTrace") 
    file_paths = [os.path.join(logStacktrace_path, f) for f in os.listdir(logStacktrace_path) if os.path.isfile(os.path.join(logStacktrace_path, f))]
    for file_path in file_paths:
        file = open(file_path)
        text = file.read()
        data.append(my_utils.tokenize_stuctured_text(text))
        
```
